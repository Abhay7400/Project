```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv') 
```

```python
#1st thing to do is i want to merge two dataset and work on single dataset *
movies=movies.merge(credits,on='title')
```

```python
#budget is not importat in recommending movie
# genres -> imp
# id -> imp use in fetching posters
# keyword -> imp
# here original language is not that imp because mostly all movies is in english language
# we keep only title because in original title there mighht chance title in other languge
# ovverview -> imp
# cast -> imp
# crew -> imp
```

```python
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.dropna(inplace=True)
```

```python
def convert11(obj):
    L = []
    for i in obj:
        L.append(i['name'])
    return L
    
    
```

```python
movies['genres'] = movies['genres'].apply(convert11)
```

```python
#convert('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')


# it's getting error because string indices must be integers

# we have to convert string of list to list
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies.head()
```

```python
movies['keywords']=movies['keywords'].apply(convert)
movies.head()
```

```python
movies.head()
```

```python
# now cast column
movies['cast'][0]
```

```python
#  want only 3 cast name  
def convert_cast(obj):
    counter = 0
    L = []
    for i in ast.literal_eval(obj):
        if counter !=3:
            L.append(i['name'])
            counter += 1
        else:
            break
    return L
```

```python
movies['cast']=movies['cast'].apply(convert_cast)
```

```python
movies.head()
```

```python
movies['crew'][0]
```

```python
#now we fetch director 
def convert_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
            break
    return L
            
    
```

```python
movies['crew']=movies['crew'].apply(convert_director)
```

```python
movies.head()
```

```python
movies['overview'][0]
```

```python
#overview is in string formate so we have to convert string into list formate to concanate with other columns
movies['overview'] = movies['overview'].apply(lambda x:x if isinstance(x, list) else [ x ])
```

```python
movies.head()
```

```python
movies['overview'][0]
```

```python
# sam worthington
# samworthington   
#remove space betweem one sentence 
def collapse(obj):
    L=[]
    for i in obj:
        L.append(i.replace(" ",""))
    return L
    
```

```python
movies['genres']=movies['genres'].apply(collapse)
movies['overview']=movies['overview'].apply(collapse)
movies['crew']=movies['crew'].apply(collapse)
movies['cast']=movies['cast'].apply(collapse)
movies['keywords']=movies['keywords'].apply(collapse)

```

```python
movies.head()
```

```python
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
new=movies.drop(columns=['overview','cast','crew','genres','keywords'])
```

```python
new
```

```python
movies['tags'].apply(lambda x:x)
```

```python
new['tags']=new['tags'].apply(lambda x:" ".join(x))
```

```python
new['tags'][0]
```

```python
new['tags']=new['tags'].apply(lambda x:x.lower())
```

```python
new['tags'][0]
```

```python
#now tag hasbeen generated now we convert tag into vector 
#after convert movies(text) into vebctor we measure similarity in tag 

from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(new['tags']).toarray()
```

```python
vector.shape
```

```python
vector
```

```python
cv.get_feature_names()
```

```python
#apply stem to remove repated words
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
#function for removing repeated words
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
    
```

```python
new['tags'] = new['tags'].apply(stem)
```

```python
#count simmiler words in tag  between movies
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vector)
```

```python
#det cosine value of words
similarity
```

```python
def recommend(movie):
    movie_index=new[new['title']==movie].index[0]
    distance=similarity[movie_index]
    movie_list=sorted(list(enumerate(distance)),reverse=True,key=lambda x:x[1])[1:10]
    for i in movie_list:
        print(new.iloc[i[0]].title)
    
    return
```

```python
recommend('Avatar')
```
